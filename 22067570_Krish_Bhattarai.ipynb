```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
os.getcwd()
```

```python
#Write a python program to load data into pandas DataFrame 
sal=pd.read_csv("DataScienceSalaries.csv") #creating a dataframe named sal
sal
```

```python
#Write a python program to remove unnecessary columns i.e., salary and salary currency.
unnecessary = ['salary', 'salary_currency']
sal = sal.drop(columns = unnecessary)
sal
```

```python
sal.columns
```

```python
#Write a python program to remove the NaN missing values from updated dataframe.
sal = sal.dropna()
sal
```

```python
#Write a python program to check duplicates value in the dataframe. 
duplicateValues = sal[sal.duplicated()]
duplicateValues
```

```python
#Write a python program to see the unique values from all the columns in the dataframe.

unique_dict = {} #Creating an empty dictionary

for column in sal.columns: #Iterating through the dataframe
    unique_dict[column] = sal[column].unique()

for column, values in unique_dict.items(): #Displaying the unique values
    print(f"Unique Values '{column}':") #print(f) to print the column name
    print(values)
    print()
```

```python
#Rename the experience level of the given columns
#SE â€“ Senior Level/Expert
sal['experience_level'] = sal['experience_level'].replace("SE","Senior Level")
sal
```

```python
#MI â€“ Medium Level/Intermediate
sal['experience_level'] = sal['experience_level'].replace("MI","Medium Level")
sal
```

```python
#EN â€“ Entry Level
sal['experience_level'] = sal['experience_level'].replace("EN","Entry Level")
sal
```

```python
#EN â€“ Executive level
sal['experience_level'] = sal['experience_level'].replace("EX","Executive Level")
```

```python
#Displaying the data of the Executive Level experience
Executive_Level = sal[sal['experience_level'] == 'Executive Level']
Executive_Level
```

```python
#Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable. 

variable = input ("Enter the Variable name: ")

if variable in sal.columns:
    summary = sal[variable].describe()
    
    skewness = sal[variable].skew()
    kurtosis = sal[variable].kurtosis()
    
    print("STATS: ")
    print(summary)
    
    print("\n")
    
    print("Other Stats: ")
    print(f"Skewness: {skewness}")
    print(f"kurtosis: {kurtosis}")
else:
    print(f"Error: The chosen variable '{variable}' doesn't exist.")
    
```

```python
#Write a Python program to calculate and show correlation of all variables.
num_sal = sal.select_dtypes(include=['number'])

cor_matrix = num_sal.corr()

print("CORRELATION MATRIX: ")
print(cor_matrix)
```

```python
#Write a python program to find out top 15 jobs. Make a bar graph of sales as well.         
top_fifteen_jobs = sal['job_title'].value_counts().head(15).index.tolist()
top_fifteen_jobs
```

```python
#Bar Graph of the sales data. (Edit the Graph)
job_counts = sal['job_title'].value_counts().head(15)
plt.figure(figsize=(10, 6))
job_counts.plot(kind='bar', color='red')  # Using barh for horizontal bar graph
plt.xlabel('Frequency')
plt.ylabel('Job Title')
plt.title('Top 15 Jobs by Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
#Which job has the highest salaries? Illustrate with bar graph.
highest_salary = sal['salary_in_usd'].max()
highest_salaried_job = sal.loc[sal['salary_in_usd'] == highest_salary, 'job_title']
highest_salaried_job
```

```python
#Write a python program to find out salaries based on experience level. Illustrate it through bar graph.
salary_based_on_exp = sal.groupby('experience_level')['salary_in_usd'].max()
salary_based_on_exp
```

```python
#Displaying it using a bar graph
plt.figure(figsize=(8, 6))
salary_based_on_exp.plot(kind='bar', color='red')
plt.title('Salaries Based on Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Maximum Salary (USD)')
plt.xticks(rotation=45)
plt.tight_layout()

```
